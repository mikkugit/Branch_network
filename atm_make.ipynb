```python
import pandas as pd
import random
```

```python
def action_type(w):
  if(w<0.5):
    return 1
  if(w>0.8):
    return -1
  else:
    return 0
```

```python
data=pd.read_csv('/content/drive/MyDrive/atm/transaction.csv')
```

```python
col=[]
for i in range(1,10001):
  col.append(i)
```

```python
df=pd.DataFrame(columns=['atm_id','location','s_amt','w_amt','bal','ration','action'])
df['atm_id']=col
```

```python
data
```

```python
locations_list =['Indiranagar','Rajajinagar','Malleswaram','Shivajinagar','Vasanth Nagar','CV Raman Nagar','Hoodi','Krishnarajapuram','Mahadevapura','Marathahalli','Varthur','Whitefield','HBR Layout','Banaswadi','Horamavu','Kalyan Nagar','Kammanahalli','Hebbal','Jalahalli']
l=len(locations_list)
for i in df.index:
  k=random.randint(0,l-1)
  df['location'][i]=locations_list[k]
  df['s_amt'][i]=random.randint(500,9000)*50
  df['w_amt'][i]=0
```

```python
for i in data.index:
  q=data['atm_id'][i]-1
  df['w_amt'][q]+=data['withdraw'][i]
```

```python
t=0
for i in df.index:
  df['bal']=df['s_amt']-df['w_amt']
  if(df['bal'][i]<0):
    t+=1
  df['ration'][i]=df['bal'][i]/df['s_amt'][i]
  df['action'][i]=action_type(df['ration'][i])

df
```
